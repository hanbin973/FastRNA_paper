```python
%load_ext memory_profiler

import numpy as np

import scipy
import scipy.sparse as sparse
import scipy.io as io

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import scanpy as sc

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut, cross_val_score
from sklearn.metrics import f1_score, silhouette_score, silhouette_samples

from openTSNE import TSNE

import matplotlib
matplotlib.rcParams['font.sans-serif'] = "Arial"

from fastrna.core import *
```

# Load data

```python
mtx = sparse.load_npz('../datasets/mouse_atlas/GSE119945_gene_count.npz').tocsc().astype(np.float32)
meta = pd.read_csv('../datasets/mouse_atlas/cell_annotate.csv')
```

# FastRNA

```python
# sort data by batch
batch_label = pd.factorize(meta.embryo_id)[0]
idx_reorder = batch_label.argsort()
blab_reorder = batch_label[idx_reorder]
ct_reorder = meta.Main_trajectory.values[idx_reorder]
mtx_reorder = mtx[:,idx_reorder]
```

```python
%time %memit gene_vars = fastrna_hvg(mtx_reorder, blab_reorder)
gene_order_var = gene_vars.argsort()[::-1]
mtx_hvg = mtx_reorder[gene_order_var[:2000],:] 
mtx_hvg.sort_indices()
numi = np.asarray(mtx_reorder.sum(axis=0)).ravel()
%time %memit eig_val, eig_vec, pca, rrt = fastrna_pca(mtx_hvg, numi, blab_reorder)
```

```python
#%config InlineBackend.figure_formats = ['svg']
fig, ax = plt.subplots(figsize=(5,4))

times = np.array([5.49, 22.20, 27.69])

sns.barplot(x=['Feature selection', 'Dimension reduction', 'FastRNA total'],
           y=times, ax=ax, edgecolor='black', linewidth=2)

# annotation
ax.set_xticklabels(ax.get_xticklabels(), fontsize=15, rotation=18, ha='right')
ax.set_ylabel('Time (s)', fontsize=18)
ax.set_yticklabels(ax.get_yticks().astype(int), fontsize=15)

ax.set_ylim(np.array(ax.get_ylim())*1.2)
ax.bar_label(ax.containers[0], fontsize=15)

plt.show()
```

# Feature selection comparison

```python
tsne = pd.read_csv('tsne_mouse_organogenesis.csv',index_col=0).values
```

```python
sum_mat = sparse.csr_matrix((np.ones(mtx_sub.shape[1]), 
                  batch_label,
                  np.arange(mtx_sub.shape[1]+1)),
                           )
gene_per_batch = mtx_sub @ sum_mat
ten_per_batch = (mtx_sub>10) @ sum_mat
```

```python
hvg_logpca = pd.read_csv('../fastrna_r/feature_var/mouse_organogenesis/var_log.csv').x
hvg_ap = pd.read_csv('feature_var/mouse_organogenesis/var_ap.csv').residual_variances.values
hvg_fr = pd.read_csv('feature_var/mouse_organogenesis/var_fr.csv')['0'].values
hvg_sct = pd.read_csv('../fastrna_r/feature_var/mouse_organogenesis/var_sct.csv').x
hvgs = [hvg_logpca, hvg_sct, hvg_ap]
```

```python
#%config InlineBackend.figure_formats = ['svg']
meta = pd.read_csv('../datasets/mouse_atlas/cell_annotate.csv')
batch_label = pd.factorize(meta.embryo_id)[0]
idx_reorder = batch_label.argsort()
ct_reorder = meta.Main_trajectory.values[idx_reorder]

fig, ax = plt.subplots(1, 4, figsize=(25, 5))
color = np.vectorize(lambda x: 2 if x else 1)(
                ((np.sort(ten_per_batch.toarray())[:,-1] / np.sort(ten_per_batch.toarray())[:,-3]) > 100))
labels = ['LogNorm', 'scTransform', 'Analytic Pearson residual']
for i, (hvg, label) in enumerate(zip(hvgs, labels)):
    
    scatter = ax[i].scatter(hvg, hvg_fr, c=color, s=10, cmap='bwr', rasterized=True)
    
    legend1 = ax[i].legend(scatter.legend_elements()[0], ['Preserved', 'Manually removed'], loc="upper left", frameon=False,
                   fontsize=15)
    ax[i].add_artist(legend1)
    
    if i == 0:
        ax[i].set_ylabel('FastRNA variance', fontsize=18)
    ax[i].set_xlabel('%s variance' % label, fontsize=18)
    
    ax[i].set_xlim([-0.5, 25.5])
    ax[i].set_ylim([-0.5, 25.5])
    
    ax[i].plot([-1,26], [-1,26], ls='--', c='grey')
    
sns.scatterplot(x=tsne[:,0], y=tsne[:,1], ax=ax[3], edgecolor='none', hue=ct_reorder, rasterized=True, s=1)
ax[3].legend().set_visible(False)
handles0, labels0 = ax[3].get_legend_handles_labels()
legend = fig.legend(handles0, 
           labels0, 
           bbox_to_anchor=(0.5,0.03), 
           loc='upper center', 
           ncol=5, 
           fontsize=18, 
           frameon=False)
ax[3].set_xlim([-35,35])
ax[3].set_ylim([-25,25])

plt.show()
```
