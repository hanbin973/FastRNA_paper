```python
%load_ext memory_profiler

import numpy as np

import scipy
import scipy.sparse as sparse
import scipy.io as io

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import scanpy as sc

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut, cross_val_score
from sklearn.metrics import f1_score, silhouette_score, silhouette_samples

from openTSNE import TSNE

from fastrna.mkl_funcs import *
from fastrna.utils import *
from fastrna.core import fastrna_hvg, fastrna_pca

import matplotlib
matplotlib.rcParams['font.sans-serif'] = "Arial"
```

```python
fr = []
ap = []
de_facLocs = [1.2, 1.5, 2, 2.5, 3, 4]
for i, de_facLoc in enumerate(de_facLocs):
    # Read splatter simulated data
    mtx = io.mmread('../fastrna_r/simdata/%s_%s_%s_%s.mtx' % (5, 2, 2000, de_facLoc)).tocsc().astype(np.float32)
    meta = pd.read_csv('../fastrna_r/simdata/meta_%s_%s_%s_%s.csv' % (5, 2, 2000, de_facLoc), index_col=0)
    gene = pd.read_csv('../fastrna_r/simdata/gene_%s_%s_%s_%s.csv' % (5, 2, 2000, de_facLoc), index_col=0)
    
    # sort data by batch
    batch_label = pd.factorize(meta.Batch)[0]
    idx_reorder = batch_label.argsort()
    blab_reorder = batch_label[idx_reorder]
    ct_reorder = meta.Group.values[idx_reorder]
    mtx_reorder = mtx[:,idx_reorder]

    # Run FastRNA HVG
    gene_vars = fastrna_hvg(mtx_reorder, blab_reorder)
    fr.append(gene_vars)

    # Run scanpy
    adata = sc.AnnData(mtx.T)
    sc.experimental.pp.highly_variable_genes(adata, flavor="pearson_residuals", n_top_genes=2000)
    ap.append(adata.var.residual_variances)
    
```

```python
import matplotlib
matplotlib.rcParams['font.sans-serif'] = "Arial"
#%config InlineBackend.figure_formats = ['svg']

fig, ax = plt.subplots(2, 3, figsize=(21,12), sharey=True)    
for i, de_facLoc in enumerate(de_facLocs):    
    # Load R packages results
    var_log = pd.read_csv('../fastrna_r/feature_var/simdata/var_log_5_2_2000_%s.csv' % de_facLoc, index_col=0).x
    var_sct = pd.read_csv('../fastrna_r/feature_var/simdata/var_sct_5_2_2000_%s.csv' % de_facLoc, index_col=0).x
    var_scry = pd.read_csv('../fastrna_r/feature_var/simdata/var_scry_5_2_2000_%s.csv' % de_facLoc, index_col=0).x
    gene = pd.read_csv('../fastrna_r/simdata/gene_%s_%s_%s_%s.csv' % (5, 2, 2000, de_facLoc), index_col=0)
    
    # python result
    var_fr = fr[i]
    var_ap = ap[i]
    
    # Plot
    x, y = int(i/3), i%3
    names = ['LogNorm', 'SCTransform', 'Scry', 'Analytic Pearson', 'FastRNA']
    var_arrs = [var_log, var_sct, var_scry, var_ap, var_fr]
    for arr, name in zip(var_arrs, names):
        gene_order_var = arr.argsort()[::-1]
        sensitivity = np.cumsum(gene.isDEG[gene_order_var].astype(int)) / gene.isDEG[gene_order_var].sum() 
        specificity = np.arange(mtx.shape[0])
        ax[x,y].plot(specificity[:5000], sensitivity[:5000], lw=3, label=name)
        ax[x,y].legend(frameon=False, fontsize=15, loc=2)
        ax[x,y].set_title('LogFC threshold: %s' % de_facLoc, fontsize=22)
        ax[x,y].set_ylim([-0.1, 1.3])
        
        ax[x,y].tick_params(axis='both', which='major', labelsize=18)
        
    fig.text(0.09, 0.5, 'Proportion of recovered DEGs', rotation=90, fontsize=22, va='center', ha='right')
    fig.text(0.5, 0.08, 'Number of selected genes', fontsize=22, va='top', ha='center')
```
